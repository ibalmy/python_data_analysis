# 亚马逊
## 简介：该数据集包含超过一千条亚马逊产品的评分、评论、价格等信息，数据来自于亚马逊官方网站上列出的详细信息。
变量含义：
- product_id ：产品id
- product_name：产品的名称

- discounted_price：产品的折扣价格
- actual_price：产品的实际价格
- discount_percentage：产品折扣的百分比
- rating：消费者的评分
- rating_count：参与评分的消费者数量
- about_product：产品的描述
- user_id：为产品撰写评论的消费者id
- user_name：为产品撰写评论的消费者名称
- review_id：评论的id
- review_title：简短的评论题目
- review_content：评论内容
- img_link：产品的图片链接
- product_link：产品的官网链接

# 清理数据

根据前面评估部分得到的结论，我们需要进行的数据清理包括：
把discounted_price 变量的数据类型转换为为数值
把actual_price 变量的数据类型转换为数值
把discount_percentage变量的数据类型转换为数值
把rating 变量的数据类型转换为数值
把rating_count 变量的数据类型转换为数值
把rating变量缺失的观察值删除

```python
cleaned_data= original_data.copy()
```

```python
cleaned_data["discounted_price"] = cleaned_data["discounted_price"] .str.extract(r'([\d\.]+)')[0].astype(float)
```

```python
cleaned_data["actual_price"] = cleaned_data["actual_price"] .str.extract(r'([\d\.]+)')[0].astype(float)
```

```python
cleaned_data["discount_percentage"]  = cleaned_data["discount_percentage"] .str.replace('%', '', regex=True).astype(float) / 100
```

```python
cleaned_data["rating_count"] = cleaned_data["rating_count"] .str.replace(',', '').astype(float)
```

```python
cleaned_data["rating"] = pd.to_numeric(cleaned_data["rating"].str.replace(',', ''), errors='coerce')
```

```python
cleaned_data=cleaned_data.rename(columns= {"discounted_price":"discounted_price(₹)","actual_price":"actual_price(₹)","discount_percentage":"discount_percentage(%)"})
```

```python
cleaned_data=cleaned_data[cleaned_data["rating_count"]>0]
```

# 保存清理后的数据

```python
cleaned_data.to_csv("amazon_cleaned.csv", index=False)
```
